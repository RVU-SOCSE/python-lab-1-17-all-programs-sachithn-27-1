```python
import pandas as pd

data = {
    'Name': ['Yoruhu', 'Sadie', 'Avery', 'Dross'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

print("DataFrame from Scratch:")
print(df)
print("\n")

df['Bonus'] = df['Salary']* 0.10

print("\n DataFrame after adding 'Bonus' column:")
print(df)
```

```python
import pandas as pd

df=pd.read_csv("McD.csv")
print(df)
print("\n")

print(df.Growth_rate_percent.mean())
print("\n")
print(df.Growth_rate_percent.median())
print("\n")
print(df.Growth_rate_percent.mode()[0])
print("\n")
print(df.Growth_rate_percent.min())
print("\n")
print(df.Growth_rate_percent.max())
print("\n")
print(df.Growth_rate_percent.var())
print("\n")
print(df.Growth_rate_percent.std())
```
