```python
# 1. Basic Try-Except
try:
    a=10
    b=0

    print(f"{a}/{b} = {a/b}")

except:
    print("Error occured!")
```

```python
# 2. Handling Specific Exceptions

try:
    n=int(input("Enter a number:"))
    print(n)

except ValueError:
    print("Invalid Input! Please enter a number.")
```

```python
# 3. Multiple Except Blocks

try: 
    a=int(input("Enter number:"))
    b=int(input("Enter number:"))
    print(f"{a}/{b} = {a/b}")

except ValueError:
    print("Invalid number entered!")
    
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
# 4. Using else block

try: 
    a=10
    b=2
    r=a/b

except ZeroDivisionError:
    print("Cannot divide by zero!")

else:
    print(f"{a}/{b}={a/b}")
```

```python
# 5. Using Finally Block

try:
    f=open("data.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found!")

finally:
    print("Execution compeleted")
```

```python
# 6. Handling Multiple Execption in One Line

try:
    n=int("abc")

except(ValueError, TypeError):
    print("Error occured da while uhm uhh coverting value.")
```

```python
# Raising an exception
try:
    age = -5
    if age < 0:
        raise ValueError("Dei you're mad or what. How come age can be negative??")
except ValueError as e:
    print(f"Error caught: {e}")
```

```python
# 8. User Defined Exception

class MyError(Exception):
    pass

try:
    raise MyError("Custom error occured")

except MyError as e:
    print(e)
```

```python
# 9. Try-Except with File Hnadling

try:
    file=open("sample.txt","r")
    content=file.read()
    print(content)

except FileNotFoundError:
    print("File does not exist!")
```
