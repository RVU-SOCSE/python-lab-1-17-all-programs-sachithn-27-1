```python
print(f"Temperature Converter: \n 1. Celsius to Fahrenheit \n 2. Fahrenheit to Celsius \n")

choice = int(input("Enter your choice (1 or 2): "))

if choice == 1:
    c = float(input("Enter temperature in Celsius: "))
    f = (c * 9/5) + 32
    print(f"{c}°C in Fahrenheit = {f}°F")

elif choice == 2:
    f = float(input("Enter temperature in Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"{f}°F in Celcius = {c}°C")

else:
    print("Invalid choice.")
```
