```python
n1=int(input(f"Enter first number: \n"))
n2=int(input(f"Enter second number: \n"))

op=input(f"Select an operator: \n 1.Addition \n 2.Subtraction \n 3.Multiplication \n 4.Division \n")

if op == "1":
    print(f"{n1} + {n2} = {n1+n2} \n")
elif op == "2":
    print(f"{n1} - {n2} = {n1-n2} \n")
elif op == "3":
    print(f"{n1} x {n2} = {n1*n2} \n")
elif op == "4":
    if n2!= "0":
        print(f"{n1} / {n2} = {n1/n2} \n")
    else:
        print ("Division by 0 not possible")
else:
    print(f"Enter valid option. \n Select 1,2,3 or 4 only. \n")
```
