```python
#Mathematical operations
import numpy as np
a = np.array([1, 2, 3])

print(a + 2)     # Add 2 to each element
print(a * 3)     # Multiply each element
```

```python
# Sum, Mean, Max, Min
import numpy as np

a=[1,6,5]
print(np.sum(a))
print(np.mean(a))
print(np.median(a))
print(np.max(a))
print(np.min(a))
```

```python
#Flattening
import numpy as np
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)
print(b)
print(b.flatten())

```

```python
#Adding & Deleting Elements
import numpy as np

a = np.array([1, 2, 3])

a = np.append(a, 4)
print(a)

a = np.delete(a, 1)   # delete index 1
print(a)
```

```python
#Sorting
import numpy as np
a = np.array([5, 2, 8, 1])
print(np.sort(a))
```

```python
import numpy as np
arr1=np.array([1,2,3,4,5,6])
arr2=np.array([10,20,30,40,50,60])
print("array 1=",arr1,"\n array 2=",arr2)

addarr=arr1+arr2 
print(addarr)
subarr=arr2-arr1
print(subarr)
mularr=arr1*arr2   
print(mularr)
```

```python
# 2D array
a1=np.array([
    [1,2,3],
    [4,5,6]
])

print(a1)
a2=np.array([
    [10,20,30],[40,50,60]
])
print(a2)

adda=a1+a2
print(adda)
mula=a1*a2   
print(mula)
```

```python
# Matrix multiplication
import numpy as np

matA=np.array([[1,2,3],[4,5,6]])
matB=np.array([[2,2,2],[3,3,3],[4,4,4]])
print("Mat A\n",matA,"\n MatB:\n",matB)
print("\n")
print(np.shape(matA))
print(np.shape(matB))

productAB=np.dot(matA,matB)
print(productAB)
```

```python
prod=np.dot(a1,a2.T)    # Transpose a2   shape 2 x 3 , 3 x 2 => 3=3
print(a1,"\n",a2)
print("transpose a2:\n")
print(a1,"\n",a2.T)
print("\nproduct")
print(prod)
```
